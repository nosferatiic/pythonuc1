```python
colecao = {}


class Carro:
    def __init__(self, marca, modelo, ano, cor, placa):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.cor = cor
        self.placa = placa
        self._ligado = False 
        self._emplacado = False
        self._seguro = False
        self._rebocado = False
        self._garantia = True

    def garantia(self):
        if 2025 - int(self.ano) <= 5:
            print('Seu carro ainda está na garantia.')
        elif 2025 - int(self.ano) >= 5:
            self._garantia = False
            print('Seu carro não está mais na garantia.')

    def ligar(self):
        if not self._ligado:
            self._ligado = True
            print(f"{self.marca} {self.modelo} está ligado.")
        else:
            print(f"{self.marca} {self.modelo} já está ligado.")

    def emplacar(self):
        if self._emplacado:
            print(f'O carro já tem placa, {self.placa}.')
        else:
            self._emplacado = True
            self.placa = input('Escolha a placa do seu carro, no modelo XXX0X00.').upper()

    def desligar(self):
        if self._ligado:
            self._ligado = False
            print(f"{self.marca} {self.modelo} está desligado.")
        else:
            print(f"{self.marca} {self.modelo} já está desligado.")

    def exibir_info(self):
        status = "ligado" if self._ligado else "desligado"
        print(f"{self.marca} {self.modelo} ({self.ano}, {self.cor}) está {status}.")
        if self._emplacado:
            print(f'Sua placa é {self.placa}.')
        else:
            print('Você não tem placa, seu fora-da-lei.')
        if self._seguro:
            print('Você tem seguro.')
        else:
            print('Você não tem seguro.')
        if self._rebocado:
            print('Você já instalou um reboque.')
        else:
            print('Você não instalou um reboque.')
        if 2025 - int(self.ano) <= 5:
            print('Seu carro ainda está na garantia.')
        else:
            print('Seu carro não está na garantia.')
       
       

    def info_cor(self):
        print(f"A cor do carro é {self.cor}.")

    def seguro_contratar(self):
        if not self._seguro:
            self._seguro = True
            print('Parabéns, você contratou um seguro.')
        else:
            print('Você já tem seguro :)')
    
    def seguro_cancelar(self):
        if self._seguro == True:
            self._seguro = False
            print('Você cancelou seu seguro :(')
        else:
            print('Você não tem seguro :(')

    def reboque_instalar(self):
        if not self._rebocado:
            self._rebocado = True
            print('Você instalou um reboque.')
        else:
            print('Você já tem reboque.')



while len(colecao) < 3:
    print('Você quer... \n 1. Ver se seu carro está na garantia \n 2. Ligar seu carro \n 3. Desligar seu carro \n 4. Emplacar seu carro \n 5. Exibir a informação do seu carro')
    print(' 6. Contratar um seguro\n 7. Cancelar um seguro\n 8. Instalar um reboque\n')
    marcauser = input('Escolha a marca: ')
    modelouser = input('Agora, o modelo: ')
    anouser = input('Seu ano: ')
    coruser = input('E a cor?')
    placauser = 'Placa'
    carrouser = Carro(marcauser, modelouser, anouser, coruser, placauser)
    print('Okay! Agora vamos tunar!')

    number = input('Escolha: ')
    while carrouser._emplacado == False or carrouser._seguro == False or carrouser._rebocado == False:
        match number:
            case '1':
                carrouser.garantia()
            case '2':
                carrouser.ligar()
            case '3':
                carrouser.desligar()
            case '4':
                carrouser.emplacar()
            case '5':
                carrouser.exibir_info()
            case '6':
                carrouser.seguro_contratar()
            case '7':
                carrouser.seguro_cancelar()   
            case '8':
                carrouser.reboque_instalar()
            case '_':
                print('Esse número é inválido.')
    if carrouser._emplacado == True and carrouser._seguro == True and carrouser._rebocado == True:
        colecao[carrouser.placa] = carrouser.exibir_info()



```
