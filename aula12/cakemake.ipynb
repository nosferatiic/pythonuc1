```python
class Bolo:
    def __init__(self, massa, calda, topping):
        self.massa = massa
        self.calda = calda
        self.topping = topping
        self._assado = False 
        self._decorado = False
        self._comido = False

    def assar(self):
        if self._comido:
            print('Você já comeu o bolo.')
        elif not self._assado:
            self._assado = True
            print(f"Seu bolo de {self.massa} está no forno... Ding!")
        elif self._assado and not self._decorado:
            print(f"Seu bolo de {self.massa} está pronto! Agora vamos decorar!")
        else:
            print('Seu bolo está assado e decorado! Hora de comer.')

    def decorar(self):
        if self._comido:
            print('Você já comeu o bolo.')
        elif not self._decorado and not self._assado:
            print('Você não pode decorar massa crua. Considere assar seu bolo.')
        elif not self._decorado and self._assado:
            self._decorado = True
            print(f"Seu bolo de {self.massa} está sendo coberto de {self.calda} e {self.topping}. \n Já pode comer!")
        else:
            print(F'Seu bolo de {self.massa} já está decorado com {self.calda} e {self.topping}. \n Come, que tá esfriando!')

    def comer(self):
        if self._comido:
            print('Você já comeu o bolo.')
        elif not self._assado:
            print('Massa crua faz mal. Considere assar seu bolo.')
        elif self._assado and not self._decorado:
            print(f'Você come a massa pura, que nem um animal selvagem.\n Estava bom o bolo de {self.massa}?')
        else:
            print(f'Você come seu delicioso bolo de {self.massa} com {self.calda} e {self.topping}. \n Tem gosto de código binário. Eca.')
            self._comido = True
 


n = 0
print('Boas vindas ao simulador de bolo!')
massauser = input('Escolha o sabor da massa: ')
caldauser = input('Agora, o da calda: ')
toppinguser = input('Por fim, seu topping: ')
bolouser = Bolo(massauser, caldauser, toppinguser)
print('Okay! Agora vamos cozinhar!')
while n == 0:
    print('Você quer... \n 1. Assar seu bolo \n 2. Decorar seu bolo \n 3. Comer seu bolo \n 4. Sair')
    number = input('Escolha: ')
    match number:
        case '1':
            bolouser.assar()
        case '2':
            bolouser.decorar()
        case '3':
            bolouser.comer()
        case '4':
            print('Obrigado por brincar comigo!')
            break
        case '_':
            print('Esse número é inválido.')
```
